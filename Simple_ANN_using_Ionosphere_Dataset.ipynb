```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/Ionosphere.csv"
df = pd.read_csv(url, sep=",")
df.head()
```

```python
# Number of independant variables = 34
X =df.drop(["Class"], axis=1)
X.head()
```

```python
# Number of dependant variable = 1 --> class column
y = df["Class"]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
pip list
```

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(units=10, activation="relu", input_dim=len(X_train.columns))) #input_dim = len(X_train) = 34
model.add(Dense(units=5, activation="relu"))
model.add(Dense(1, activation="sigmoid")) # Output layer | sigmoid-->probabilistic value
model.summary
```

```python
model.compile(optimizer="sgd", loss="binary_crossentropy", metrics=["accuracy"])
#binary_crossentropy --> bianary classification
#sgd --> Stochastic Gradient Descent
# No of iterartions = 9 = total no of columns in training dataset/batchsize
```

```python
model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=100, batch_size=32)
```

```python
model.save("ionosphere.h5")
```
